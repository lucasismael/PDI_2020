```python
import cv2
import numpy as np
import math
from matplotlib import pyplot as plt
%matplotlib inline
```

## Leitura de Imagem:

```python
img = cv2.imread(r'C:/Users/Lucas Ismael/Documents/PDI_2020/images/animacao.png')

plt.figure(figsize=(10, 10))
plt.axis("off")
plt.imshow(img), plt.title('Imagem original - Espa√ßo BGR', fontsize=20, fontweight="bold")
```

* Por padr√£o o OpenCV utiliza o espa√ßo de cores BGR, por isso a imagem sai azulada.
* Para corrigir esse problema convertemos o espa√ßo de cores BGR para RGB.

```python
img_corrigida = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 10))
plt.axis("off")
plt.imshow(img_corrigida), plt.title('Imagem original corrigida - Espa√ßo RGB', fontsize=20, fontweight="bold")
```

```python
blueEQ = cv2.equalizeHist(bluechannel)
greenEQ = cv2.equalizeHist(greenchannel)
redEQ = cv2.equalizeHist(redchannel)

redchannelhisteq = np.zeros(255, int) #Initializing intensity vector
greenchannelhisteq = np.zeros(255, int) #Initializing intensity vector
bluechannelhisteq = np.zeros(255, int) #Initializing intensity vector

for h in range(len(red_eq)):
    redchannelhisteq[h] = np.count_nonzero(blueEQ == h)
    greenchannelhisteq[h] = np.count_nonzero(greenEQ == h)
    bluechannelhisteq[h] = np.count_nonzero(blueEQ == h)

plt.figure(figsize=(20, 10))
plt.subplot(2,3,1)
plt.axis("off")
plt.imshow(redEQ, cmap = 'Reds'), plt.title('Canal vermelho - Equalizado', fontsize=14, fontweight="bold")
plt.subplot(2,3,2)
plt.axis("off")
plt.imshow(greenEQ, cmap = 'Greens'), plt.title('Canal verde - Equalizado', fontsize=14, fontweight="bold")
plt.subplot(2,3,3)
plt.axis("off")
plt.imshow(blueEQ, cmap = 'Blues'), plt.title('Canal azul - Equalizado', fontsize=14, fontweight="bold")
plt.subplot(2,3,4)
plt.axis("on")
plt.plot(redchannelhisteq), plt.title('Imagem original - Canal vermelho', fontsize=14, fontweight="bold")
plt.subplot(2,3,5)
plt.axis("on")
plt.plot(greenchannelhisteq), plt.title('Imagem original - Canal verde', fontsize=14, fontweight="bold")
plt.subplot(2,3,6)
plt.axis("on")
plt.plot(bluechannelhisteq), plt.title('Imagem original - Canal azul', fontsize=14, fontweight="bold")
```

```python
EqualizedImage = cv2.merge((blueEQ, greenEQ, redEQ))

plt.figure(figsize=(20, 20))
plt.subplot(1,2,1)
plt.axis("off")
plt.imshow(img_corrigida), plt.title('Imagem original', fontsize=20, fontweight="bold")
plt.subplot(1,2,2)
plt.axis("off")
plt.imshow(cv2.cvtColor(EqualizedImage, cv2.COLOR_BGR2RGB)), plt.title('Imagem equalizada', fontsize=20, fontweight="bold")
```

## Experimento 2 - Transformando o espa√ßo de cores BGR para HSV

```python
imgHSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

(huechannel, saturationchannel, brightnesschannel) = cv2.split(imgHSV)

plt.figure(figsize=(20, 20))
plt.subplot(1,4,1)
plt.axis("off")
plt.imshow(imgHSV), plt.title('Imagem no espa√ßo HSV', fontsize=18, fontweight="bold")
plt.subplot(1,4,2)
plt.axis("off")
plt.imshow(huechannel), plt.title('Canal matiz', fontsize=18, fontweight="bold")
plt.subplot(1,4,3)
plt.axis("off")
plt.imshow(saturationchannel), plt.title('Canal satura√ß√£o', fontsize=18, fontweight="bold")
plt.subplot(1,4,4)
plt.axis("off")
plt.imshow(brightnesschannel), plt.title('Canal valor/brilho', fontsize=18, fontweight="bold")
```

## Experimento 3 - Equalizando a imagem no espa√ßo HSV
