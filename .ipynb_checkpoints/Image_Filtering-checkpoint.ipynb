```python
import subprocess, os
import numpy as np
import matplotlib.pyplot as plt
import cv2
import random
from PIL import Image, ImageFilter
```

# Filtragem: 

## Adição de ruído sal e pimenta:

```python
def my_spnoise(*args):
    noise_output = np.zeros(img.shape, np.uint8)
    #prob = float(input('a: '))
    prob = 0.1
    thres = 1-prob
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            rdn = random.random()
            if rdn < prob:
                noise_output[i][j] = 0
            elif rdn > thres:
                noise_output[i][j] = 255
            else:
                noise_output[i][j] = img[i][j]
    return noise_output

img = cv2.imread(r'C:/Users/Lucas Ismael/Documents/PDI_2020/images/'+input('Selected input(.pgm):')+'.pgm', 1)
noise_image = my_spnoise(img)

plt.figure(figsize=(10, 10))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(noise_image),plt.title('Imagem com ruído')
plt.xticks([]), plt.yticks([])
plt.show()
```

## Filtro Laplaciano:

```python
img = cv2.imread(r'C:/Users/Lucas Ismael/Documents/PDI_2020/images/'+input('Selected input(.pgm):')+'.pgm')

laplacian5 = cv2.Laplacian(img,cv2.CV_64F, ksize = 5)


plt.figure(figsize=(20, 20))
plt.subplot(141),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(142),plt.imshow(laplacian5),plt.title('Laplaciano 5x5')
plt.xticks([]), plt.yticks([])
plt.subplot(143),plt.imshow(img + laplacian5),plt.title('Imagem + Laplaciano 5x5')
plt.xticks([]), plt.yticks([])
plt.subplot(144),plt.imshow(img - laplacian5),plt.title('Imagem - Laplaciano 5x5')
plt.xticks([]), plt.yticks([])
plt.show()
```
